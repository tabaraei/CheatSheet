```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

sns.set_style('darkgrid')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/tabaraei/CheatSheet/master/datasets/mpg.csv')
sns.distplot(df[df['cyl'] == 4].hwy)
plt.show()
```

# Scatter Plot

```python
plt.figure(figsize=(7, 7))
ax = sns.scatterplot(data=df, x='displ', y='hwy', hue='cyl', size='cyl', palette='Set2') # palette GnBu_d, Set1, Set3, magma, spring, cubehelix
plt.show()
```

# Bar Chart:

```python
plt.figure(figsize=(8, 4))
ax = sns.barplot(data=df, x='class', y='cty', palette='Set3', ci=None, estimator=np.mean)
plt.show()
```

# Box Plot:

```python
plt.figure(figsize=(12, 8))
ax = sns.boxplot(data=df, x='class', y='hwy', palette='Set3', showfliers=True)
plt.show()
```

# Heat Map:

```python
plt.figure(figsize=(8, 6))
ax = sns.heatmap(df.corr(),annot=True)
plt.show()
```
